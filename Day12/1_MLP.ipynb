```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.metrics import accuracy_score, log_loss
from sklearn.neural_network import MLPClassifier
import os
```

```python
os.chdir("D:\\meridianthe4\\PML\\Cases\\Wisconsin")
```

```python
wisconsin = pd.read_csv("BreastCancer.csv")
```

```python
le = LabelEncoder()
wisconsin['Class'] = le.fit_transform(wisconsin['Class'])
X, y = wisconsin.drop('Class', axis=1), wisconsin['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25, stratify=y)
```

```python
scl_X = MinMaxScaler()
X_train_scaled = scl_X.fit_transform(X_train)
X_test_scaled = scl_X.transform(X_test)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,), activation='logistic', random_state=25, max_iter=500)
mlp.fit(X_train_scaled, y_train)
y_pred_proba = mlp.predict_proba(X_test_scaled)
print("Log Loss:", log_loss(y_test, y_pred_proba))
y_pred = mlp.predict(X_test_scaled)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
