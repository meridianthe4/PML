```python
import pandas as pd
import re
import numpy as np
```

```python
dataset = pd.read_csv("Restaurant_Reviews.tsv", sep='\t')
dataset.head()
```

```python
corpus = []
for i in range(0, dataset.shape[0]):
    review = dataset['Review'][i]
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    review = review.split()
    corpus.append(review)
```

```python
len(corpus)
```

```python
from gensim.models import Word2Vec
model_r = Word2Vec(corpus, min_count=1, sg=1, vector_size=300)
```

### Generating the sentence embeddings

```python
means = []
for sentence in corpus:
    word_vects = [model_r.wv.get_vector(word, norm=True) for word in sentence]
    row_means = np.mean(word_vects, axis=0)
    means.append(row_means)
X = np.array(means)
y = dataset['Liked']
```
