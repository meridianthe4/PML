```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier, VotingRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression, Ridge, ElasticNet
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, log_loss
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.metrics import precision_score, accuracy_score, f1_score, mean_squared_error, mean_absolute_error
from tqdm import tqdm
import os
os.chdir("D:\\meridianthe4\\PML\\Datasets")
```

```python
boston = pd.read_csv("Boston.csv")
X, y = boston.drop(columns=["medv"]), boston["medv"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
ridge = Ridge()
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)
mae_ridge = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (Ridge):", mae_ridge)
```

```python
el = ElasticNet()
el.fit(X_train, y_train)
y_pred_el = el.predict(X_test)
mae_el = mean_absolute_error(y_test, y_pred_el)
print("Mean Absolute Error (ElasticNet):", mae_el)
```

```python
dtr = DecisionTreeRegressor(random_state=25)
dtr.fit(X_train, y_train)
y_pred_dtr = dtr.predict(X_test)
mae_dtr = mean_absolute_error(y_test, y_pred_dtr)
print("Mean Absolute Error (Decision Tree Regressor):", mae_dtr)
```

```python
voting = VotingRegressor(estimators=[
    ('ridge', ridge),
    ('elasticnet', el),
    ('decision_tree', dtr)
])
voting.fit(X_train, y_train)
y_pred_voting = voting.predict(X_test)
mae_voting = mean_absolute_error(y_test, y_pred_voting)
print("Mean Absolute Error (Voting Regressor):", mae_voting)
```

```python
voting = VotingRegressor(estimators=[
    ('ridge', ridge),
    ('elasticnet', el),
    ('decision_tree', dtr)
], weights=[8, 2, 10])
voting.fit(X_train, y_train)
y_pred_voting = voting.predict(X_test)
mae_voting = mean_absolute_error(y_test, y_pred_voting)
print("Mean Absolute Error (Voting Regressor):", mae_voting)
```
