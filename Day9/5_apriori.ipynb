```python
import pandas as pd
from mlxtend.frequent_patterns import association_rules, apriori
import os
```

```python
os.chdir("D:\\meridianthe4\\PML\\Datasets")
```

```python
fp_df = pd.read_csv("Faceplate.csv", index_col=0)
fp_df = fp_df.astype(bool)
itemsets = apriori(fp_df, min_support=0.2, use_colnames=True)
rules = association_rules(itemsets, metric="confidence", min_threshold=0.6)
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules.sort_values(['lift', 'confidence'], ascending=[False, False])
```

```python
def generate_rules(min_support, min_confidence):
    itemsets = apriori(fp_df, min_support=min_support, use_colnames=True)
    rules = association_rules(itemsets, metric="confidence", min_threshold=min_confidence)
    rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
    return rules.sort_values(['lift', 'confidence'], ascending=[False, False])

import ipywidgets as widgets
widgets.interact(generate_rules, min_support=(0.1, 0.5, 0.05), min_confidence=(0.5, 1.0, 0.05))
```
