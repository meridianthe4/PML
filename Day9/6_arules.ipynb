```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
import os
os.chdir("D:\\meridianthe4\\PML\\Datasets")
```

```python
groceries = []
with open("Groceries.csv", "r") as f:groceries = f.read()
groceries = groceries.split("\n")
groceries_list = []
for i in groceries:
    groceries_list.append(i.split(","))
groceries_list[:5]
```

```python
te = TransactionEncoder().set_output(transform="pandas")
fp_df = te.fit(groceries_list).transform(groceries_list)
```

```python
def generate_rules(min_support, min_confidence):
    itemsets = apriori(fp_df, min_support=min_support, use_colnames=True)
    rules = association_rules(itemsets, metric="confidence", min_threshold=min_confidence)
    rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
    return rules.sort_values(['lift', 'confidence'], ascending=[False, False])

import ipywidgets as widgets
widgets.interact(generate_rules, min_support=(0.001, 0.5, 0.001), min_confidence=(0.5, 1.0, 0.01))
```
