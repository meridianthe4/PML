## Boston Dataset
1. Train test split
2. Test MAE for degree 2, 3, 4

```python
boston = pd.read_csv("Boston.csv")
```

```python
X = boston.drop('medv', axis=1)
y = boston['medv']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False).set_output(transform="pandas")
X_poly = poly.fit_transform(X_train)
lr = LinearRegression()
lr.fit(X_poly, y_train)
y_pred = lr.predict(poly.transform(X_test))
root_mean_squared_error(y_test, y_pred)
```

```python
poly = PolynomialFeatures(degree=3, include_bias=False).set_output(transform="pandas")
X_poly = poly.fit_transform(X_train)
lr = LinearRegression()
lr.fit(X_poly, y_train) 
y_pred = lr.predict(poly.transform(X_test))
root_mean_squared_error(y_test, y_pred)
```

```python
poly = PolynomialFeatures(degree=4, include_bias=False).set_output(transform="pandas")
X_poly = poly.fit_transform(X_train)
lr = LinearRegression()
lr.fit(X_poly, y_train)
y_pred = lr.predict(poly.transform(X_test))
root_mean_squared_error(y_test, y_pred)
```

```python
housing = pd.read_csv("Housing.csv")
housing
```

```python
cat_housing = housing[housing.columns[housing.dtypes == 'object']]
num_housing = housing[housing.columns[housing.dtypes != 'object']]
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform="pandas")
ohe_housing = ohe.fit_transform(cat_housing)
```

```python
df_housing = pd.concat([num_housing, ohe_housing], axis=1)
```

```python
X = df_housing.drop('price', axis=1)
y = df_housing['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
root_mean_squared_error(y_test, y_pred)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False).set_output(transform="pandas")
X_poly = poly.fit_transform(X_train)
lr = LinearRegression()
lr.fit(X_poly, y_train)
y_pred = lr.predict(poly.transform(X_test))
root_mean_squared_error(y_test, y_pred)
```

```python
poly = PolynomialFeatures(degree=3, include_bias=False).set_output(transform="pandas")
X_poly = poly.fit_transform(X_train)
lr = LinearRegression()
lr.fit(X_poly, y_train)
y_pred = lr.predict(poly.transform(X_test))
root_mean_squared_error(y_test, y_pred)
```

```python
poly = PolynomialFeatures(degree=4, include_bias=False).set_output(transform="pandas")
X_poly = poly.fit_transform(X_train)
lr = LinearRegression()
lr.fit(X_poly, y_train)
y_pred = lr.predict(poly.transform(X_test))
root_mean_squared_error(y_test, y_pred)
```

### Using ColumnTransformer

```python
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_selector
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform="pandas")
col_transformer = ColumnTransformer([("OHE", ohe, list(housing.columns[housing.dtypes == 'object']))], 
                                    remainder='passthrough', 
                                    verbose_feature_names_out=False)
col_transformer = col_transformer.set_output(transform="pandas")
col_transformer.fit_transform(housing)
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform="pandas")
col_transformer = ColumnTransformer([("OHE", ohe, make_column_selector(dtype_include=object))], 
                                    remainder='passthrough', 
                                    verbose_feature_names_out=False)
col_transformer = col_transformer.set_output(transform="pandas")
col_transformer.fit_transform(housing)
```
