```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_predict, cross_val_score, KFold
from sklearn.metrics import r2_score
import os
```

```python
os.chdir("D:\\meridianthe4\\PML\\Datasets")
```

```python
pizza = pd.read_csv("pizza.csv")
X = pizza[['Promote']]
y = pizza['Sales']
```

```python
lr = LinearRegression()
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
# next(iter(kfold.split(pizza)))
```

```python
for i, (train_index, test_index) in enumerate(kfold.split(pizza)):
    print(f"Fold {i}")
    print(f" Train: index = {train_index}")
    print(f" Test: index = {test_index}")
```

```python
scores = []
for i, (train_index, test_index) in enumerate(kfold.split(pizza)):
    print(f"Fold {i}")
    print(f" Train: index = {train_index}")
    print(f" Test: index = {test_index}")
    
    X_train, y_train = X.iloc[train_index], y.iloc[train_index]
    X_test, y_test = X.iloc[test_index], y.iloc[test_index]
    lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)
    scores.append([r2_score(y_test, y_pred)])
```

```python
print(f"CV Scores = {np.mean(scores)}")
```

```python
results = cross_val_score(lr, X, y, cv=kfold, scoring='r2')
print(f"Scores = {results}")
print(f"CV Scores = {np.mean(scores)}")
```
