```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import os
os.chdir('d:/meridianthe4/PML')
```

```python
milk = pd.read_csv('D:\meridianthe4\PML\Datasets\Milk.csv',index_col=0)
```

```python
from sklearn.decomposition import PCA
# milk = pd.read_csv('Milk.csv',index_col=0) # to maintain the index col as it is 
scaler = StandardScaler().set_output(transform = 'pandas')
milk_scaled = scaler.fit_transform(milk)
prcomp = PCA().set_output(transform='pandas')
pca_data = prcomp.fit_transform(milk_scaled)
pca_data.var()
```

```python
prcomp.explained_variance_
```

```python
prcomp.explained_variance_ratio_* 100
```

```python
from pca import pca
import matplotlib.pyplot as plt
```

```python
model = pca()
results = model.fit_transform(milk_scaled,col_labels=milk.columns,row_labels=milk.index)
model.biplot(label=True,legend=True)
for i in np.arange(0,milk.shape[0]):
    plt.text(pca_data.values[i,0],pca_data.values[i,1],milk.columns[i],list[milk.index][i],color='red')

plt.show()
```
