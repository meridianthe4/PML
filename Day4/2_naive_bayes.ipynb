```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB, GaussianNB
from sklearn.metrics import log_loss
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.impute import SimpleImputer
import os
```

```python
os.chdir("D:\\meridianthe4\\PML\\Cases\\Cancer")
```

```python
data = pd.read_csv("Cancer.csv")
```

```python
le = LabelEncoder()
data['Class'] = le.fit_transform(data['Class'])
X, y = data.drop(columns=['Class']), data['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=25)
```

```python
# imputer = SimpleImputer(strategy='mean')
# X_train_imputed = imputer.fit_transform(X_train)
# X_test_imputed = imputer.transform(X_test)
```

```python
# ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform="pandas")
ohe = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore').set_output(transform="pandas")
X_train_ohe = ohe.fit_transform(X_train)
X_test_ohe = ohe.transform(X_test)
```

```python
nb = BernoulliNB()
nb.fit(X_train_ohe, y_train)
y_pred_prob = nb.predict_proba(X_test_ohe)
print(f"Log Loss: {log_loss(y_test, y_pred_prob)}")
y_pred = nb.predict(X_test_ohe)
print(f"Accuracy: {(y_pred == y_test).mean()}")
```

```python
glass = pd.read_csv("Glass.csv")
```

```python
le = LabelEncoder()
glass['Type'] = le.fit_transform(glass['Type'])
X, y = glass.drop(columns=['Type']), glass['Type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_prob = nb.predict_proba(X_test)
print(f"Log Loss: {log_loss(y_test, y_pred_prob)}")
y_pred = nb.predict(X_test)
print(f"Accuracy: {(y_pred == y_test).mean()}")
```

```python
hr = pd.read_csv("D:\\meridianthe4\\PML\\Cases\\HR_comma_sep.csv")
X, y = hr.drop('left', axis=1), hr['left']
ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform="pandas")
col_transformer = ColumnTransformer([("OHE", ohe, make_column_selector(dtype_include=object))], 
                                    remainder='passthrough', 
                                    verbose_feature_names_out=False)
col_transformer = col_transformer.set_output(transform="pandas")
X = col_transformer.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_prob = nb.predict_proba(X_test)
print(f"Log Loss: {log_loss(y_test, y_pred_prob)}")
y_pred = nb.predict(X_test)
print(f"Accuracy: {(y_pred == y_test).mean()}")
```
