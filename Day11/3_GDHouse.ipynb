```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_absolute_error
```

```python
os.chdir('D:\\meridianthe4\\PML\\Datasets')
```

```python
boston = pd.read_csv('Boston.csv')
X  = boston.drop('medv', axis=1)
y = boston[['medv']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
scl_x , scl_y = MinMaxScaler(), MinMaxScaler()
X_trn_scl , y_trn_scl = scl_x.fit_transform(X_train), scl_y.fit_transform(y_train.values.reshape(-1,1))
X_tst_scl = scl_x.transform(X_test)
sgd = SGDRegressor(random_state=25) 
sgd.fit(X_trn_scl, y_trn_scl)
y_pred_scl = sgd.predict(X_tst_scl)
y_pred = scl_y.inverse_transform(y_pred_scl.reshape(-1,1))
mean_absolute_error(y_test, y_pred) 
# mean_absolute_error
```

```python
etas = [0.001, 0.01, 0.1, 0.2, 0.4]
lr_sch = ['constant', 'optimal', 'adaptive','invscaling']
scores = []
for e in etas:
    for s in lr_sch:
        sgd = SGDRegressor(learning_rate=s, eta0=e, random_state=25) 
        sgd.fit(X_trn_scl, y_trn_scl)
        y_pred_scl = sgd.predict(X_tst_scl)
        y_pred = scl_y.inverse_transform(y_pred_scl.reshape(-1,1))
        scores.append([e,s,mean_absolute_error(y_test, y_pred)])

df_scores = pd.DataFrame(scores, columns=['eta','sch','score'])
df_scores.sort_values('score')
```
