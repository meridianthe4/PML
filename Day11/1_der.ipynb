```python
import numpy as np
```

```python
def f(x):
    return np.exp(x) - np.log(x)
```

```python
def fd(x):
    return np.exp(x) - 1/x
```

```python
x = 1.5              
learning_rate = 0.1    
iterations = 50     
tolerance = 1e-6       

print(f"Starting at x = {x}, f(x) = {f(x)}")

for i in range(iterations):
    gradient = fd(x)
    new_x = x - learning_rate * gradient

    if abs(new_x - x) < tolerance:
        print(f"Converged at iteration {i}")
        break
        
    x = new_x
    if i % 10 == 0:
        print(f"Iteration {i}: x = {x:.5f}, f(x) = {f(x):.5f}")

print(f"\nFinal Result:")
print(f"Minimum found at x = {x:.5f}")
print(f"Minimum value f(x) = {f(x):.5f}")
```
